```python
def is_location_enabled(location):
    if pd.isna(location) or location == 0 or location == 'False' or location == '':
        return False
    return True

# Create column 'enable_location'
df_profile['enable_location'] = df_profile['location'].apply(is_location_enabled)
```

```python
df_selected = df_profile[['ID', 'protected',
       'followers_count', 'friends_count', 'listed_count', 'favourites_count',
       'statuses_count', 'created_at', 'geo_enabled', 'enable_location','verified', 'default_profile',
       'default_profile_image', 'domain', 'label',
       'account_age_years',
       'tweets_count']]

columns_to_convert = ['followers_count', 'friends_count', 'listed_count', 'favourites_count',
       'statuses_count','account_age_years',
       'tweets_count']  

# Convert each column
for col in columns_to_convert:
    df_selected[col] = pd.to_numeric(df_selected[col], errors='coerce')
```

```python
df_merged = pd.merge(df_selected, df_tweets_mean, on='ID', how='left')

# Replace NaN values with 0 at the embeddings columns
embedding_columns = [col for col in df_merged if col.startswith('tweet_')]
df_merged[embedding_columns] = df_merged[embedding_columns].fillna(0)
```

## Visualize some differences between bot and humans accounts

```python
label_mapping = {0: 'human', 1: 'bot'}
df_relation_copy = df_relation.copy()
df_relation_copy['label'] = df_relation_copy['label'].map(label_mapping)
```

```python
fig = px.histogram(df_relation_copy, x='verified', color='label', barmode='overlay', nbins=30)

fig.update_layout(
    title='Verified Count Distribution',
    xaxis_title='verified_count',
    yaxis_title='Count',
    width=800,    
    height=400    
)

fig.show()
```

Most of the 'verified' accounts are humans, so we need to be careful with this feature and perform different experiments to assess whether including or excluding the 'verified' variable significantly biases the results.

```python
fig = px.histogram(df_relation_copy, x='account_age_years', color='label', barmode='overlay', nbins=30)

fig.update_layout(
    title='Account age years Distribution',
    xaxis_title='Account_age_years',
    yaxis_title='Count',
    width=800,   
    height=400   
)


fig.show()
```

- Seams like bots are associated with newer accounts (created more recently).

```python
fig = px.histogram(df_relation_copy, x='geo_enabled', color='label', barmode='overlay', nbins=30)

# Atualiza o layout para melhor visualização
fig.update_layout(
    title='Geo enabled Distribution',
    xaxis_title='geo_enabled',
    yaxis_title='Count',
    width=800,    
    height=400    
)

fig.show()
```

- This term typically refers to a setting that allows Twitter to collect and use precise location data from a device. When a tweet is geo-enabled, it means that the tweet can include specific location information, such as coordinates (latitude and longitude), that is associated with the device at the time of tweeting.
- So, apparently, most bot accounts do not enable geo location.

```python
fig = px.histogram(df_relation_copy, x='enable_location', color='label', barmode='overlay', nbins=30)


fig.update_layout(
    title='Location Distribution',
    xaxis_title='Location',
    yaxis_title='Count',
    width=800,   
    height=400    
)


fig.show()
```

- This feature refers to settings related to whether a user allows Twitter to attach general location information to their tweets. When enabled, Twitter can add a general location (such as a city or neighborhood) to the user's tweets.
- Unlike geo-enabled tweets, which use precise coordinates, this feature uses broader location data and is typically less specific. 
- The 'enable_location' feature doesn't seem to be as relevant as 'geo_enable,' since the proportion of True and False for both humans and bots is similar."
